```python
import re
import pandas as pd

# Charger le fichier texte
with open('/Users/ines/NLP/qa_en_1.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Découper par bloc id=
blocs = re.split(r'\nid=\d+', '\n' + text)
data = []

for bloc in blocs:
    if not bloc.strip():
        continue
    # Extraire l'URL
    url = re.search(r'(https?://[^\s]+)', bloc)
    url = url.group(1) if url else ''
    # Extraire la Description/Question
    question = re.search(r'Description\s*(.*?)(?:\nDialogue)', bloc, re.DOTALL|re.IGNORECASE)
    question = question.group(1).strip() if question else ''
    # Extraire le texte du Patient (optionnel)
    patient = re.search(r'Patient:\s*(.*?)(?:Doctor:)', bloc, re.DOTALL|re.IGNORECASE)
    patient = patient.group(1).strip() if patient else ''
    # Extraire la réponse du Doctor
    answer = re.search(r'Doctor:\s*(.+?)(?=\n\s*\n|$)', bloc, re.DOTALL|re.IGNORECASE)
    answer = answer.group(1).strip() if answer else ''
    data.append({
        "url": url,
        "question": question,
        "patient": patient,
        "answer": answer,
    })

# Conversion en DataFrame
df = pd.DataFrame([d for d in data if d['question'] and d['answer']])
print(df[['question','answer']].head(5))
df.to_csv('meddialog_clean.csv', index=False)
```

```python
#Triage dataset

import pandas as pd

# Lecture du dataset PRET pour le ML
df = pd.read_csv('/Users/ines/NLP/triage_dataset.csv')

print(df.columns)
print(df.head())

#Suppression lignes vides
df = df.dropna(subset=['question', 'triage'])  # ou les colonnes pertinentes

#Suppression doublons
df = df.drop_duplicates(subset=['question', 'triage'])  # ou juste 'question'

#Filtre sur la longueur
df = df[df['question'].str.len() > 10]

#Suppression si class n'est pas valide 
df = df[~df['triage'].isnull()]

#Nettoyage basique de texte
df['question'] = df['question'].str.strip()


# Si déjà clean > tu passes direct à l'audit :
print("Nombre de lignes :", len(df))
print("Noms de colonnes :", df.columns.tolist())

# Exemples d’audit (repris du script précédent) :
if 'triage' in df.columns:
    print("Nombre de classes :", df['triage'].nunique())
    print("Distribution :\n", df['triage'].value_counts())
else:
    print("Aucune colonne 'triage' trouvée.")

print("Longueur moyenne question :", df['question'].str.len().mean())

print("Questions uniques :", df['question'].nunique())

```

```python
import pandas as pd
import random

sexes = ["F", "M"]
medical_conditions = [
    "Hypertension", "Asthma", "Type II diabetes", "History of stroke",
    "Seasonal allergies", "No significant medical history", "Coronary artery disease", "Heart failure",
    "Osteoarthritis", "Crohn’s disease", "Obesity", "Hypercholesterolemia",
    "Breast cancer", "Prostate cancer", "Depression", "Early-stage Alzheimer’s"
]
recent_history_pool = [
    "Fever and sore throat for the past 3 days", "Recent chest pain, dry cough",
    "Sudden weakness in right arm, speech difficulties", "Sneezing, runny nose, red eyes",
    "Fall at home without loss of consciousness", "Acute back pain radiating to the leg",
    "Loss of appetite, nausea and vomiting", "Persistent cough, shortness of breath on exertion",
    "Recent anxiety attack", "Chronic fatigue", "Non-healing skin wounds",
    "Unexplained night sweats", "Painless hematuria", "Generalized itching",
    "Severe headaches, photophobia", "Dizziness with exertion", "Insomnia, sleep disturbances",
    "Nocturnal breathing discomfort", "Sudden exhaustion while walking", "Occasional palpitations"
]

data = []
for i in range(1001, 1101):
    age = random.randint(18, 89)
    sex = random.choice(sexes)
    if random.random() > 0.2:
        medical_history = ", ".join(random.sample(medical_conditions, k=random.randint(1, 3)))
    else:
        medical_history = "No significant medical history"
    recent_history = random.choice(recent_history_pool)
    data.append({
        "patient_id": i,
        "age": age,
        "sex": sex,
        "medical_history": medical_history,
        "recent_history": recent_history
    })

df = pd.DataFrame(data)
df.to_csv("patient_history_100.csv", index=False)
print("Sample data:\n", df.head())
print("File 'patient_history_100.csv' saved.")
```
